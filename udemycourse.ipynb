# Udemy Course Dataset Analysis

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('UdemyCoursesDataset.csv')
data.head(2)
```

# Q1. What are different subjects for which udemy is offering course ?

```python
data.subject.unique()
```

# Q2. Which subject has maximum number of courses

```python
data.subject.value_counts().sort_values(ascending=False).head(1)
```

# Q3. Show All Courses which are free of cost.

```python
data[(data.is_paid == False) & (data.price == "Free")]
```

# Q4. Show All courses which are paid.

```python
data[(data.is_paid == True) & (data.price != "Free")]
```

# Q5. Which are top selling courses.

```python
data.sort_values(by="num_subscribers", ascending=False).head(10)
```

# Q6. Which are least selling courses.

```python
data.sort_values(by="num_subscribers").head(10)
```

# Q7. Show All courses of graphic designing where price is below 100.

```python
data[(data.subject == "Graphic Design") & (data.price < "100")]
```

# Q8. List all of the courses which are related with 'Python'

```python
data[data.course_title.str.contains("python")]
```

# Q9. What are the courses that are published in year '2015'

```python
data.published_timestamp = pd.to_datetime(data.published_timestamp)
data[data.published_timestamp.dt.year == 2015]
```

# Q10. List Max number of subscribers for each level of course

```python
data.groupby("level")["num_subscribers"].max()
```
